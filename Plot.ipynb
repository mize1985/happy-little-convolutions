```python
import pandas as pd
import seaborn as sns
```

```python
artists_df = pd.read_csv('./multiTimeline.csv')
artists_df['Month'] = pd.to_datetime(artists_df['Month'])
artists_df.set_index('Month', inplace=True)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,12))
plt.plot(artists_df['Claude Monet: (Worldwide)'].rolling(window=12).mean(), linewidth = 3, label = 'Claude Monet')

plt.plot(artists_df['Vincent van Gogh: (Worldwide)'].rolling(window=12).mean(), linewidth = 3, label = 'Vincent van Gogh')
plt.plot(artists_df['Bob Ross: (Worldwide)'].rolling(window=12).mean(), linewidth = 3, label = 'Bob Ross')
plt.legend(fontsize=15)
plt.title("Painters' Search Frequency Over Time", fontsize=35)
plt.xlabel('Date', fontsize=30)
plt.ylabel('Worldwide Relative Search Volume', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('plot.png')
```
